```python
#recursive function
def fact(n):
    if n==0:

     return 1
    else:
        return n*fact(n-1)
print(fact(5))
```

```python
#function for max of 3 numbers
def max(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c
print(max(10,20,30))

    
```
