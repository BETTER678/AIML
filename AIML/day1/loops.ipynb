```python
#factorial
n = int(input("enter the number"))
fact= 1
for i in range (1,n+1):
    fact=fact*i
print(fact)
```

```python
#Armstrong number
def is_armstrong(n):
    original=n
    sum=0
    while n>0:
        digit=n%10
        sum+=digit**3
        n=n//10
    return original==sum
print(is_armstrong(153))
print(is_armstrong(123))
```

```python
#Prime number check
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
print(is_prime(7))
print(is_prime(10))
```
